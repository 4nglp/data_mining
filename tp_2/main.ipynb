importing pandas, matplotlib and seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

setting the dataframe

```python
df = pd.read_csv("../tp_1/Housing.csv")
```

the distribution of the houses' price

```python
plt.figure()
sns.histplot(df["price"], kde=True)
plt.show()
# we can see that we have a positive skew that means that mean > med > mode
```

 the distribution of the areas

```python
plt.figure()
sns.histplot(data=df["area"],kde=True)
plt.show()
# another positive skew
```

the distribution of both bed and bath rooms' number

```python
r = [df["bedrooms"],df["bathrooms"]]
plt.figure(figsize=(10,5))
for i in range(len(r)):
    plt.subplot(1,2,i+1)
    sns.histplot(data=r[i],kde=30)
plt.show()
```

the relationship between the area and the price

```python
plt.figure()
sns.regplot(x=df["area"],y=df["price"],scatter=True)
plt.show()
# we can say that we have a positive relationship where as the area increase the pice does as weel
```

 the relationship between the number of bedrooms and the price

```python
plt.figure()
sns.boxenplot(x=df["bedrooms"],y=df["price"])
plt.show()
```

the relationship between the airconditioning and the price

```python
plt.figure()
sns.boxplot(x=df["airconditioning"],y=df["price"])
plt.show()
# we can see that both the mean and median on the yes part are bigger than the no part
# which can let us say that the price gets higher if there's an airconditioning system installed
```

the impact of the furnishing status on the price

```python
plt.figure(figsize=(15,5))
sns.boxplot(x=df["furnishingstatus"],y=df["price"])
plt.show()
# so as we can see if the house is furnished it tends to have a higher price where furnished > semi > none
# then we can say that the furnishing status does indeed have a positve impact on the price

```

the overall or global correlation

```python
corr = df.corr(numeric_only=True)
plt.figure()
sns.heatmap(corr,annot=True)
plt.show()
# the closer to 1 the stronger the correlation weither is it positve or negative
```

the variables most correlated with price

```python
corr_price = df.corr(numeric_only=True)["price"].drop("price").sort_values(ascending=False)
print(corr_price)
# add a bar plot using plt
plt.figure()
sns.barplot(corr_price)
plt.show()
```
