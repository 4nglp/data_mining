# TP N° 1 : Découverte du Dataset, Analyse Statistique Descriptive et EDA (Exploratory Data Analysis)

importing pandas and loading the dataset

```python
import pandas as pd

df = pd.read_csv("Housing.csv")
```

1st 10 lines

```python
print(df.head(10))
```

the dataset's shape (lines x cols)

```python
print(df.shape)
```

cols of the dataset 

```python
for col in df.columns:
    print(col)
```

types of data

```python
print(df.dtypes)
```

calculating the number of int and object data types

```python
print(df.dtypes.value_counts())
```

the dataset's general state

```python
print(df.info())
```

null values count

```python
print(df.isnull().sum())
print("no null values")
```

global statistics

```python
print(df.describe())
```

the 1st and 3rd quantiles of the dataset

```python
print(df.quantile([0.25,0.75],numeric_only=True))
```

mean and median of each house's price

```python
mean = df["price"].mean()
print("mean: ",mean.round(2))
median = df["price"].median()
print("median: ",median.round(2))
```

distribution of cats for each categorial var

```python
df_objects = df.select_dtypes(include=['object'])
for col in df_objects.columns:
    print(df_objects[col].value_counts())
```

null vals in each col

```python
print(df.isnull().sum())
```

top 5 areas with the highest price mean

```python
df_means = df.groupby("area")["price"].mean()
print(df_means.sort_values(ascending=False).head(5))
```

mean price based by furnishing status

```python
print(df.groupby("furnishingstatus")["price"].mean().round(2))
```

mean price of houses that have airconditioning, basement and on the main road

```python
print("airconditioning:", df[df["airconditioning"] == "yes"]["price"].mean().round(2))
print("basement:", df[df["basement"] == "yes"]["price"].mean().round(2))
print("mainroad:", df[df["mainroad"] == "yes"]["price"].mean().round(2))
```

mean price of the houses based on the stories number

```python
print(df.groupby("stories")["price"].mean().round(2).sort_values(ascending=False))
```
